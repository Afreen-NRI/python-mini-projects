```python
def easy_quiz():
    score = 0
    
    # Question 1
    print("Question 1: What is 2* 2?")
    answer1 = input("a) 3 b) 4 c) 5 d) 6 Your answer: ")

    if answer1 == "b":
        print("Correct")
        score += 10
    else:
        print("Incorrect. The correct answer is 4.")
    
    # Question 2
    print("Question 2: What is 3 * 2?")
    answer2 = input("a) 10 b) 9  c) 6 d) 13 Your answer: ")
    
    if answer2 == "c":
        print("Correct")
        score += 10
    else:
        print("Incorrect. The correct answer is 6.")

        # Question 3
    print("Question 3: What is 5 * 3?")
    answer3 = input("a) 6 b) 7 c) 15 d) 9 Your answer: ")
    
    if answer3 == "c":
        print("Correct")
        score += 10
    else:
        print("Incorrect. The correct answer is 15.")

        # Question 4
    print("Question 4: What is 10 * 6?")
    answer4 = input("a) 60 b) 17 c) 15 d) 9 Your answer: ")
    
    if answer4 == "a":
        print("Correct")
        score += 10
    else:
        print("Incorrect. The correct answer is 60.")

        # Question 5
    print("Question 5: What is 5 * 5?")
    answer5 = input("a) 14 b) 7 c) 25 d) 19 Your answer: ")
    
    if answer5 == "c":
        print("Correct")
        score += 10
    else:
        print("Incorrect. The correct answer is 25.")

        # Question 6
    print("Question 6: What is 2 * 8?")
    answer6 = input("a) 45 b) 16 c) 40 d) 36 Your answer: ")
    
    if answer6 == "b":
        print("Correct")
        score += 10
    else:
        print("Incorrect. The correct answer is 16.")

        # Question 7
    print("Question 7: What is 3 * 9?")
    answer7 = input("a) 23 b) 27 c) 38 d) 28 Your answer: ")
    
    if answer7 == "b":
        print("Correct")
        score += 10
    else:
        print("Incorrect. The correct answer is 27.")
    
    # Question 8
    print("Question 8: What is 4 * 4?")
    answer8 = input("a) 60 b) 70 c) 16 d) 90 Your answer: ")
    
    if answer8 == "c":
        print("Correct")
        score += 10
    else:
        print("Incorrect. The correct answer is 16.")

        # Question 9
    print("Question 9: What is 6 * 6?")
    answer9 = input("a) 36 b) 197 c) 198 d) 190 Your answer: ")
    
    if answer9 == "a":
        print("Correct")
        score += 10
    else:
        print("Incorrect. The correct answer is 36.")

        # Question 10
    print("Question 10: What is 10 * 10?")
    answer10 = input("a) 60 b) 100 c) 74 d) 76 Your answer: ")
    
    if answer10 == "b":
        print("Correct")
        score += 10
    else:
        print("Incorrect. The correct answer is 100.")
    
    
    # Final score
    print(f"Your final score is: {score}/100")

# Run the quiz
easy_quiz()
```
